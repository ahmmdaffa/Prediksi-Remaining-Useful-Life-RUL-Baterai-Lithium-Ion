# 01 - Pengambilan Data & Inspeksi Awal
Notebook ini membaca `baterai.csv`, menampilkan struktur dataset, missing values, dan statistik deskriptif awal.

```python
from pathlib import Path
import warnings

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from IPython.display import Markdown, display

warnings.filterwarnings("ignore")
sns.set_theme(style="whitegrid", context="notebook")
pd.set_option("display.max_columns", 50)

DATA_PATH = Path("baterai.csv")
if not DATA_PATH.exists():
    raise FileNotFoundError(f"File tidak ditemukan: {DATA_PATH.resolve()}")

# Membaca dataset menggunakan pandas.
df = pd.read_csv(DATA_PATH)

print(f"Ukuran dataset: {df.shape[0]:,} baris dan {df.shape[1]:,} kolom")
print("\nDaftar kolom:")
print(df.columns.tolist())
print("\nInformasi tipe data:")
df.info()

display(df.head())
display(df.isna().sum().to_frame("Jumlah Missing Value"))
display(df.describe())
```

**Analisis Hasil**

Dataset berhasil dimuat menggunakan pandas. Dataset berisi data sensor baterai lithium-ion, yaitu `Battery_ID`, `Cycle_Number`, `Time_Seconds`, `Voltage_V`, `Current_A`, `Temperature_C`, dan `Capacity_Ah`. Inspeksi awal dilakukan untuk memahami ukuran data, tipe data, missing values, serta rentang nilai numerik. Tahap ini penting karena kualitas data awal memengaruhi validitas proses EDA, preprocessing, modelling, dan evaluasi model prediksi RUL.
